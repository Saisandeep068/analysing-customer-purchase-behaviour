```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file into a DataFrame
df = pd.read_excel(r'C:\Users\RGUKTS\Downloads\online+retail\Online_Retail.xlsx')

# Convert the date column to datetime format
df['Date'] = pd.to_datetime(df['InvoiceDate'])

# Group by date and calculate total sales, using absolute value of Quantity
sales_over_time = df.groupby('InvoiceDate')['Quantity'].apply(lambda x: x.abs().sum()).reset_index()

# Create a line chart
plt.figure(figsize=(10,8))
plt.plot(sales_over_time['InvoiceDate'], sales_over_time['Quantity'])
plt.xlabel('Date')
plt.ylabel('Quantity')
plt.title('Sales over Time')
plt.show()
```

```python
# Import necessary libraries
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Select relevant columns
customer_data = df[['CustomerID', 'Quantity', 'UnitPrice']].copy()

# Scale the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(customer_data[['Quantity', 'UnitPrice']])

# Cast scaled data to appropriate dtype
customer_data['Quantity'] = scaled_data[:, 0].astype(float)
customer_data['UnitPrice'] = scaled_data[:, 1].astype(float)

# Perform K-Means clustering
kmeans = KMeans(n_clusters=5)
customer_data['Cluster'] = kmeans.fit_predict(customer_data[['Quantity', 'UnitPrice']])

# Print the cluster assignments
print(customer_data['Cluster'])

```

```python
# Import necessary libraries
import matplotlib.pyplot as plt

# Create a scatter plot
plt.scatter(customer_data['Quantity'].abs(), customer_data['UnitPrice'], c=customer_data['Cluster'])
plt.xlabel('Quantity')
plt.ylabel('UnitPrice')
plt.title('Customer Purchase Behavior Clusters')
plt.show()
```
